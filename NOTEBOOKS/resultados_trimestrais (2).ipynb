```python
import pandas as pd
path = 'C:/Users/vinicius_coin/Desktop/DADOS_MESTRADO/bovesparser-master/data/'
base_consolidada = 'PREBASE/'
```

```python
# Carrega os registros do ITR_DRE do último exercício dos três primeiros trimestres
# DRE - Demonstração de Resultado
# CD_CONTA 3.05 - Resultado Antes do Resultado Financeiro e dos Tributos ou Resultado Operacional  (LAJIR/EBIT)                           

import pandas as pd
df_dre_ebit = pd.read_csv(path+base_consolidada+"DRE_consolidado.csv", sep = ';',encoding="ISO-8859-1")
df_dre_ebit = df_dre_ebit[df_dre_ebit["ORDEM_EXERC"] == "ÚLTIMO"]
contas = ['Resultado Operacional','Resultado Antes do Resultado Financeiro e dos Tributos']
df_dre_ebit = df_dre_ebit[df_dre_ebit['DS_CONTA'].isin(contas)]
df_dre_ebit.rename(columns={"VL_CONTA": "VLR_EBIT"}, inplace=True)
df_dre_ebit['VLR_EBIT'] = df_dre_ebit.loc[df_dre_ebit['ESCALA_MOEDA'].isin(['MIL','MILHAR']), ['VLR_EBIT']] * 1000
df_dre_ebit.head()
      
```

```python
# Carrega os registros do ITR_DRE do último exercício dos três primeiros trimestres
# DRE - Demonstração de Resultado
# CD_CONTA 3.13 - Lucro/Prejuízo do Período       (lucro liquido)
                          

import pandas as pd
df_dre_lucro = pd.read_csv(path+base_consolidada+"DRE_consolidado.csv", sep = ';',encoding="ISO-8859-1")
df_dre_lucro = df_dre_lucro[df_dre_lucro["ORDEM_EXERC"] == "ÚLTIMO"]
contas = ['Lucro/Prejuízo do Período']
df_dre_lucro = df_dre_lucro[df_dre_lucro['DS_CONTA'].isin(contas)]
df_dre_lucro.rename(columns={"VL_CONTA": "VLR_LUCRO_LIQUIDO"}, inplace=True)
df_dre_lucro['VLR_LUCRO_LIQUIDO'] = df_dre_lucro.loc[df_dre_lucro['ESCALA_MOEDA'].isin(['MIL','MILHAR']), ['VLR_LUCRO_LIQUIDO']] * 1000


print(df_dre_lucro.shape[0])
#df_dre_lucro.head(20)
      
```

```python
df_dre = pd.merge(df_dre_ebit, df_dre_lucro, how = 'left', left_on=['CNPJ_CIA','DT_REFER'], right_on=['CNPJ_CIA','DT_REFER'])
df_dre.head()
```

```python
# Carrega os registros do ITR_BPP do último exercício dos três primeiros trimestres
# BPP - Balanço Patrimonial Passivo
# CD_CONTA 2.05 - Patrimônio Líquido
                         
df_bpp = pd.read_csv(path+base_consolidada+"BPP_consolidado.csv", sep = ';',encoding="ISO-8859-1")
df_bpp = df_bpp[df_bpp["ORDEM_EXERC"] == "ÚLTIMO"]
contas = ['Patrimônio Líquido']
df_bpp = df_bpp[df_bpp['DS_CONTA'].isin(contas)]
df_bpp.rename(columns={"VL_CONTA": "VLR_PATRIMONIO_LIQUIDO"}, inplace=True)
df_bpp['VLR_PATRIMONIO_LIQUIDO'] = df_bpp.loc[df_bpp['ESCALA_MOEDA'].isin(['MIL','MILHAR']), ['VLR_PATRIMONIO_LIQUIDO']] * 1000

```

```python
a = df_bpp[(df_bpp["DENOM_CIA"] == 'BCO BRASIL S.A.') & (pd.to_datetime(df_bpp["DT_REFER"]) >= pd.to_datetime('2018-01-01'))]
a.head(10)
```

```python
# Carrega os registros do ITR_DVA do último exercício dos três primeiros trimestres
# DVA -  Demonstração de Valor Adicionado
# CD_CONTA 7.04.01 - Depreciação, Amortização e Exaustão
                         

import pandas as pd
df_dva = pd.read_csv(path+base_consolidada+"DVA_consolidado.csv", sep = ';',encoding="ISO-8859-1")
df_dva = df_dva[df_dva["ORDEM_EXERC"] == "ÚLTIMO"]
contas = ['Depreciação, Amortização e Exaustão']
df_dva = df_dva[df_dva['DS_CONTA'].isin(contas)]
df_dva.rename(columns={"VL_CONTA": "VLR_DEPRECIACAO_AMORTIZACAO"}, inplace=True)
df_dva['VLR_DEPRECIACAO_AMORTIZACAO'] = df_dva.loc[df_dva['ESCALA_MOEDA'].isin(['MIL','MILHAR']), ['VLR_DEPRECIACAO_AMORTIZACAO']] * 1000
df_dva.head()
```

```python
#Junta os diferentes dataframes com informações do ITR
new_df = pd.merge(df_dre, df_bpp, how = 'left', left_on=['CNPJ_CIA','DT_REFER'], right_on=['CNPJ_CIA','DT_REFER'])
df_result_itr = pd.merge(new_df, df_dva, how = 'left', left_on=['CNPJ_CIA','DT_REFER'], right_on=['CNPJ_CIA','DT_REFER'])
#df_result = df_result[['CNPJ_CIA','DT_REFER','DT_INI_EXERC','DT_FIM_EXERC','DS_CONTA_x','VL_CONTA_x','DS_CONTA_y','VL_CONTA_y','DS_CONTA','VL_CONTA']]
df_result_itr = df_result_itr[['CNPJ_CIA','DT_REFER','DT_INI_EXERC_x','DT_FIM_EXERC_x','VLR_EBIT','VLR_LUCRO_LIQUIDO','VLR_PATRIMONIO_LIQUIDO','VLR_DEPRECIACAO_AMORTIZACAO']]
df_result_itr['VLR_EBITDA'] = df_result_itr['VLR_EBIT'] - df_result_itr['VLR_DEPRECIACAO_AMORTIZACAO']
df_result_itr.head()
```

```python
df_ticker = pd.read_csv(path+"APOIO/tickers.csv", sep = ';', encoding="utf-8", usecols = ['Tickers','CNPJ']) 
df_result_itr = pd.merge(df_result_itr, df_ticker, how = 'left', left_on='CNPJ_CIA', right_on='CNPJ')
df_result_itr = df_result_itr.assign(tickers = df_result_itr['Tickers'].str[:4])
sr = pd.Series(pd.to_datetime(df_result_itr["DT_REFER"]))
df_result_itr = df_result_itr.assign(trimestre = sr.dt.quarter)
df_result_itr = df_result_itr.assign(ano = sr.dt.year)

df_result_itr.head()
```

```python
a = df_result_itr[(df_result_itr["tickers"] == 'PETR') & (pd.to_datetime(df_result_itr["DT_REFER"]) >= pd.to_datetime('2018-08-01'))]
a.head(10)
```

```python
df_cotacoes = pd.read_csv(path+"PREBASE/COTACOES_consolidado.csv", sep = ';',encoding="ISO-8859-1")

df_cotacoes = df_cotacoes.assign(cod_papel_ini = df_cotacoes['cod_papel'].str[:4])
df_cotacoes.sort_values(by=['cod_papel','data_pregao'], inplace=True)
df_cotacoes['preco_dia_anterior'] = df_cotacoes['preco_ultimo'].shift(periods=1)
x = 'a'
df_cdi = pd.read_csv(path+"APOIO/taxa_cdi.csv", sep = ';',encoding="ISO-8859-1")
df_cdi['ano_mes'] = pd.to_datetime(df_cdi['data']).dt.to_period('M') 
df_cotacoes['ano_mes'] = pd.to_datetime(df_cotacoes['data_pregao']).dt.to_period('M') 

df_cotacoes = pd.merge(df_cotacoes, df_cdi, how = 'left', left_on=['ano_mes'], right_on=['ano_mes'])
                                                                     
for index, row in df_cotacoes.iterrows():
    if row['cod_papel'] != x:
        df_cotacoes.at[index,'preco_dia_anterior'] = None        
    x = row['cod_papel']

    


df_cotacoes.head(50)
```

```python
df_cotacoes['retorno_diario_ativo'] = (df_cotacoes['preco_ultimo'] - df_cotacoes['preco_dia_anterior'])/df_cotacoes['preco_dia_anterior']*100


df_cotacoes['mean_retorno'] = df_cotacoes.groupby(['ano_mes','cod_papel'])['retorno_diario_ativo'].transform('mean')
df_cotacoes['std_retorno'] = df_cotacoes.groupby(['ano_mes','cod_papel'])['retorno_diario_ativo'].transform('std')
#df_cotacoes['sharpe_ratio'] = (df_cotacoes['mean_retorno']-pd.to_numeric(df_cotacoes['cdi']))/df_cotacoes['std_retorno']
#ver_depois


df_cotacoes.head(30)
```

```python
df_capital_social = pd.read_excel(path+"APOIO/capital_social.xlsx", sheet_name="acoes")
#df_capital_social = df_capital_social.reindex(df.index.drop(0)).reset_index(drop=True)
df_capital_social = df_capital_social[['Código','Qtde total de ações']]

df_capital_social.head()
```

```python
df1 = pd.merge(df_cotacoes, df_capital_social, how = 'left', left_on='cod_papel_ini', right_on='Código')
df1.head(30)
```

```python
sr = pd.Series(pd.to_datetime(df1["data_pregao"]))
df1 = df1.assign(trimestre = sr.dt.quarter)
df1 = df1.assign(ano = sr.dt.year)
new_df = pd.merge(df1, df_result_itr, how = 'left', left_on=['cod_papel_ini','ano','trimestre'], right_on=['tickers','ano','trimestre'])
new_df.head(30)
```

```python
df_setorial = pd.read_csv(path+"APOIO/setorial_B3.csv", usecols = ['SETOR_ECONOMICO','SUBSETOR','SEGMENTO_SETOR','CODIGO'], sep=';')
new_df = pd.merge(new_df, df_setorial, how = 'left', left_on='cod_papel_ini', right_on='CODIGO')
new_df.head(30)
```

```python
new_df['GAF'] = new_df['VLR_EBITDA']/new_df['VLR_EBIT']
new_df['ROE'] = new_df['VLR_LUCRO_LIQUIDO'] / new_df['VLR_PATRIMONIO_LIQUIDO']
new_df['LPA'] = new_df['VLR_LUCRO_LIQUIDO'] / new_df['Qtde total de ações']
new_df['VPA'] = new_df['VLR_PATRIMONIO_LIQUIDO'] / new_df['Qtde total de ações']
new_df['PL'] = new_df['preco_ultimo']/new_df['LPA'] 

new_df = new_df[['data_pregao','cod_papel','nome_resum','SETOR_ECONOMICO','SUBSETOR','SEGMENTO_SETOR','preco_ultimo','VLR_EBIT','VLR_EBITDA','VLR_LUCRO_LIQUIDO','VLR_PATRIMONIO_LIQUIDO','VLR_DEPRECIACAO_AMORTIZACAO','GAF','ROE','LPA','VPA','PL']]
# adicionar 'sharpe_ratio' acima

```
